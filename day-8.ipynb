```python
# Define a class called BopIt
class BopIt:

    # __init__ runs automatically whenever a new object is created
    def __init__(self, condition):
        # self refers to the specific object being created
        # we store the condition value inside the object
        self.condition = condition

    # Method that describes the object
    def describe(self):
        # print() separates items with spaces when commas are used
        print("This Bop It is in", self.condition, "condition.")


# Create two BopIt objects
# This runs __init__ and assigns condition = "near mint"
bop1 = BopIt("near mint")

# This runs __init__ and assigns condition = "scratched"
bop2 = BopIt("scratched")


# Use the objects by calling their describe() method
# self becomes bop1 inside the method
bop1.describe()

# self becomes bop2 inside the method
bop2.describe()
```

```python
class PlayerCharacter:

    def __init__(self, name, hp, shield_block):
        self.name = name
        self.hp = hp
        self.shield_block = shield_block

    def block(self, damage):
        # reduce damage by shield value
        reduced_damage = damage - self.shield_block

        # damage can't go below 0
        if reduced_damage < 0:
            reduced_damage = 0

        self.hp -= reduced_damage

        print(self.name, "blocks with their shield!")
        print("Incoming damage:", damage)
        print("Shield blocks:", self.shield_block)
        print("Damage taken:", reduced_damage)
        print(self.name, "has", self.hp, "HP left.\n")


class Monster:

    def __init__(self, species, hp):
        self.species = species
        self.hp = hp

    def block(self, damage):
        # monster ignores all damage
        reduced_damage = 0

        print(self.species, "blocks perfectly!")
        print("Incoming damage:", damage)
        print("Damage taken:", reduced_damage)
        print(self.species, "has", self.hp, "HP left.\n")


# create characters
hero = PlayerCharacter("Aria", 100, 8)
goblin = Monster("Goblin", 30)

incoming_damage = 20

# both use block(), but behavior differs
hero.block(incoming_damage)
goblin.block(incoming_damage)
```

using the above object classes, define a function that lets me input an actor type ('PC' or 'monster'), and the relevant stats as a list. 'PC' types should then use the list to make a player character object, and monster types should use the list to make a monster object

`maker_func('PC',['shrek',10000,400])
maker_func('monster',['clown',400])`

```python
# Class representing a player character
class PlayerCharacter:

    # Runs automatically when a new PlayerCharacter object is created
    def __init__(self, name, hp, shield_block):
        self.name = name              # store the character's name
        self.hp = hp                  # store the character's health
        self.shield_block = shield_block  # how much damage the shield blocks

    def block(self, damage):
        # reduce damage by shield value
        reduced_damage = damage - self.shield_block

        # damage can't go below 0
        if reduced_damage < 0:
            reduced_damage = 0

        # subtract the remaining damage from hp
        self.hp -= reduced_damage

        # print combat information
        print(self.name, "blocks with their shield!")
        print("Incoming damage:", damage)
        print("Shield blocks:", self.shield_block)
        print("Damage taken:", reduced_damage)
        print(self.name, "has", self.hp, "HP left.\n")


# Class representing a monster
class Monster:

    # Runs automatically when a new Monster object is created
    def __init__(self, species, hp):
        self.species = species    # monster type
        self.hp = hp              # monster health

    def block(self, damage):
        # monster ignores all damage
        reduced_damage = 0

        # print combat information
        print(self.species, "blocks perfectly!")
        print("Incoming damage:", damage)
        print("Damage taken:", reduced_damage)
        print(self.species, "has", self.hp, "HP left.\n")


# FACTORY FUNCTION
# This function creates either a PlayerCharacter or Monster object
# based on the actor_type input
def maker_func(actor_type, stats):

    # if actor_type is PC, create a PlayerCharacter
    if actor_type == 'PC':
        # unpack the stats list into variables
        name, hp, shield_block = stats

        # return a new PlayerCharacter object
        return PlayerCharacter(name, hp, shield_block)

    # if actor_type is monster, create a Monster
    elif actor_type == 'monster':
        # unpack the stats list
        species, hp = stats

        # return a new Monster object
        return Monster(species, hp)

    else:
        # basic error message if type is wrong
        print("Invalid actor type")
        return None


# create characters using the factory function
hero = maker_func('PC', ['Aria', 100, 8])
goblin = maker_func('monster', ['Goblin', 30])


incoming_damage = 20

# both use block(), but behavior differs
hero.block(incoming_damage)
goblin.block(incoming_damage)
```
